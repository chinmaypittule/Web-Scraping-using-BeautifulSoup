```python
topics_string = 'Topics\n\nArtificial Intelligence\nCareer Advice\nComputer Vision\nData Engineering\nData Science\nLanguage Models\nMachine Learning\nMLOps\nNLP\nNews\nProgramming\nPython\nSQL'

# Split the string by newline characters and filter out empty strings
topics_list = [topic.strip().lower().replace(' ', '-') for topic in topics_string.split('\n') if topic.strip()]

# Remove the first element ('Topics') since it's not a topic
topics_list = topics_list[1:]

# Print the result
print(topics_list)

```

```python
topics_list
```

```python
headings = []
for i in topics_list:
    page = requests.get(f"https://www.kdnuggets.com/tag/{i}")

    soup1 = bs(page.content,"html.parser")
    for item in soup1.find_all('h1'):
        headings.append(item.text) 
```
