```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
job_title = []
company_name = []
experience = []
location= []
keyskills = []
date_posted = []
more_details = []
```

```python
for i in range(1,10):
    page = requests.get(f"https://www.timesjobs.com/candidate/job-search.html?from=submit&luceneResultSize=25&txtKeywords=python&postWeek=60&searchType=personalizedSearch&actualTxtKeywords=python&searchBy=0&rdoOperator=OR&pDate=I&sequence={i}&startPage=1")

    soup = bs(page.content, "html.parser")

    jobs = soup.find_all('li', class_='clearfix job-bx wht-shd-bx')

    for x in jobs:
        title = x.h2.text.strip()
        job_title.append(title)
        companyname = x.find('h3').text.strip().replace('\r\n     (More Jobs)','')
        company_name.append(companyname)
        exp = x.find('li').text.replace('card_travel','')
        experience.append(exp)
        loc = x.find_all('li')[1].text.strip().replace('location_on\n','')
        location.append(loc)
        skills = x.find('span',class_='srp-skills').text.strip().replace(' ','')
        keyskills.append(skills)
        posted= x.find('span',class_='sim-posted').text.strip()
        date_posted.append(posted)
        web=x.h2.a['href']
        more_details.append(web)
    
```

```python
v = [job_title,company_name, experience, location, keyskills, date_posted, more_details]

for name, value in zip(["job_title","company_name", "experience", "location", "keyskills", "date_posted", "more_details"], v):
    print("Length of", name, ":", len(value))
```

```python
df = pd.DataFrame({"job_title":job_title,"company_name":company_name,"experience":experience,"location":location,"keyskills":keyskills,"date_posted":date_posted,"more_details":more_details})
```
